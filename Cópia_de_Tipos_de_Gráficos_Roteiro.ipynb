```python
def Graficodispersao(style):
  #estilo de cor
  plt.style.use(style) 

  # Aumentar o tamanho da fonte
  plt.rcParams.update({'font.size': 8})

  # Criar um gráfico de linha
  plt.figure(figsize=(8, 6))

  plt.scatter(df['Idade'], df['Preço'])

  plt.title('Preço do Carro x Idade')
  plt.xlabel('Idade do Carro (Anos)')
  plt.ylabel('Preço (R$)')

  
  plt.savefig('Graficobarra.svg')
  # Mostrar o gráfico
  plt.show()

```

# Qual é a diferença entre o gráfico de linha na interpretação?
R: O gráfico de linha é muito utilizado para visualizar tendências e a evolução de uma variável. Em relação ao gráfico de dispersão, suas diferenças está na natureza das variáveis representadas e na interpretação das relações entre elas.
Ou seja, o gráfico de linha é utilizado mais para a visualização de tendência e padrões ao longo do tempo, ou sequência ordenada, por exemplo. Enquanto o gráfico de dispersão é mais usado e adequado para identificar correlações, padrões de dispersão e tambem valores discrepantes, que são mais "fora" em relação as outras variáveis contínuas. Os dois tipos de gráficos são de grande ajuda para compreender e entender as informações que esta se colocando ali, e para suas utilizações dependem do contexto e objetivo que se está querendo obter.

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.normal(0, 100, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

# Interprete a distribuição distribuição e compare a diferença com a distribuição normal

R: A diferença entre os dois gráficos está, de maneira geral, na forma de distribuição das idades. No gráfico de distribuição aleatória, há uma distribuição discreta entre as idades. O gráfico esta apresentando em barras separadas a frequência de ocorrência de cada valor inteiro. No gráfico de distribuição normal tem-se uma distribuição contínua e simétrica. Nesse caso, essa distribuição é caracterizada por uma média e  um desvio padrão. A maioria das idades está em torna da média e a frequência vai diminuindo à medida que se afasta da média.O primeiro  representa uma distribuição discreta com dados aleatórios de valor inteiro, enquanto o segundo, possui uma distribuição contínua e simétrica seguindo uma distribuição normal.

# Gráfico de Donut

```python
def Graficodistribuicao(style):
    quantidade = [650, 550]
    porcentagemF = round((650 * 100) / 1200, 1)
    porcentagemM = round((550 * 100) / 1200, 1)
    porcentFstring = str(porcentagemF)
    porcentMstring = str(porcentagemM)
    generos = ['Feminino - ' + porcentFstring + '%', 'Masculino - ' + porcentMstring + '%']
    plt.style.use(style)
    
    # Criar o gráfico de pizza
    fig, ax = plt.subplots()
    pie = ax.pie(quantidade, labels=generos, startangle=90, colors=['#FFC0CB', '#0000ff'], wedgeprops={'width': 0.4})
    ax.set_title('Distribuição de Gênero')
    
    # Configurar o fundo do gráfico como preto
    fig.set_facecolor('#05092E')
    plt.title('Distribuição de Gênero', color='white')
    
    # Configurar a cor do texto "Feminino" e da porcentagem para branco
    for text in pie[1]:
        text.set_color('white')
    
    # Mostrar o gráfico
    plt.show()

Graficodistribuicao('seaborn-whitegrid')
```

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R:O grafico de distribuição aprensentado acima ajuda a ter uma melhor visualização dos dados distribuidos, permitindo-nos avaliar sua variabilidade e a presença de valores discrepante. Através dele, podemos identifica a mediana dos valores inseridos.
A linha inferior da caixa é o Q1, e a linha superior é o Q3. A caixa em si representa a dispersão dos valores centrais do conjunto de dados.
As linhas verticais que se estendem a partir da caixa são conhecidas como mostram a extensão da distribuição.Se houvesse valores fora desse intervalo, estes seriam representados por pontos individuais do lodo de fora.

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

# Explique o gráfico acima, qual é a diferença da Equipe C da Equipe D.
R: Para a representação do gráfico acima, foi utilizado um gráfico violino para se ter uma melhor visualização da distribuição das equipes a partir do conjunto de dados.De forma geral, é um gráfico de densidade,que permite analisar a distribuição dos dados de forma mais completa.

Cada violino exibe a estimativa da densidade de probabilidade do dado em diferentes partes do intervalo. A diferença entre as equipes, é que a equipe C possui uma distribuição mais ampla, com uma caixa mais larga, indicando uma maior variabilidade na pontuação. Além disso, apresenta tambem uma densidade de pontuação maior em valores mais baixos, onde pode-se dizer que indica que eles obtiveram pontuações mais baixas com maior frequência em comparação à equipe D.
